```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

```python
from florentino.linear_model import LinearRegression
```

```python
# Coefficients for x1^3, x1^2, x1, and intercept
w = np.array([-1.4, 10, -1.5, -5])
e = 100  # Noise
N = 200

low = -5
high = 10

x1 = np.random.uniform(low, high, (N, 1))
X = np.hstack([x1**3, x1**2, x1, np.ones((N, 1))])
y = (X @ w).reshape(N, 1) + e * np.random.randn(N, 1)
print(X.shape, y.shape)
```

```python
plt.scatter(x1, y, color='green', label='Generated data (Quadratic)')
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Generated Polynomial Dataset (Quadratic)')
plt.legend()
plt.show()
```

```python
model = LinearRegression(x1, y, 4)
x_test = np.linspace(low, high, 400).reshape(-1, 1)
predictions = model.predict(x_test)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x_test, predictions, color='red', label='Regression line')
plt.scatter(x1, y, color='blue', label='Training data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
