```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

```python
w = 1.5
b = 5
e = 1

np.random.seed(0)
x_train = 4 * np.random.rand(100, 1)
y_train = w * x_train + b + e * np.random.randn(100, 1)
print(x_train.shape, y_train.shape)
```

```python
plt.scatter(x_train, y_train, color='blue', label='Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
from florentino.linear_model import LinearRegression
```

```python
model = LinearRegression(x_train, y_train)
x_test = np.linspace(0, 4, 100).reshape(-1, 1)
predictions = model.predict(x_test)

plt.plot(x_test, predictions, color='red', label='Regression line')
plt.scatter(x_train, y_train, color='blue', label='Training data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
